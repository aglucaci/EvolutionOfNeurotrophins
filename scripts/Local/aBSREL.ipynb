```python
import json
import PyQt5
from ete3 import Tree, TreeStyle, PhyloTree
```

```python
file = "BDNF_codons_renamed.fasta_codon_macse.fas.ABSREL.json"
```

```python
with open(file) as json_handle :
    json_data = json.load(json_handle)
#end with
```

```python
TREE = json_data["input"]["trees"]["0"]
t = Tree(str(TREE)+";",  format=1)
ts = TreeStyle()
ts.mode = "c" # use circular mode 
#t.render("t2_image.png", tree_style=ts)
t.show(tree_style=ts)
```

```python
# load tree from newick
t2 = PhyloTree(TREE.replace(";", " ") + ";", format=1)

# Root
root_point = t2.get_midpoint_outgroup()
t2.set_outgroup(root_point)

# Styling
ts = TreeStyle()
#ts.show_leaf_name = True
ts.mode = "c" # use circular mode 

# Show the tree
#t2.show(tree_style=ts)


t2.render("t2_image.png", tree_style=ts)
```
